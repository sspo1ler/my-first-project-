```python
from csv import reader
applestore=list(reader(open('AppleStore.csv')))
google=list(reader(open('googleplaystore.csv')))
```

```python
def explore_data(dataset, start, end, rows_and_columns=True):
    dataset_slice = dataset[start:end]
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
```

```python
reviews_max = {}

for app in google[1:]:
    name = app[0]
    n_reviews = float(app[3])

    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews

    elif name not in reviews_max:
        reviews_max[name] = n_reviews
```

```python

android_clean=[]
already_added=[]


for app in google[1:]:
    name = app[0]
    n_reviews = float(app[3])

    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
def english(string):
    ff=0
    for x in string:
        if ord(x)>127:
          ff+=1
    return ff<=3



```

```python
englishappsgoogle=[]
englishappsapple=[]
for row in android_clean:
    if english(row[0]):
        englishappsgoogle.append(row)
for row in applestore[1:]:
    if english(row[1]):
        englishappsapple.append(row)
explore_data(englishappsgoogle, 0, 3, True)
print('\n')
explore_data(englishappsapple, 0, 3, True)
```

```python
applefin=[] #11
googlefin=[] #genre in 9
for row in englishappsapple:
    price = (row[4])
    if price =='0.0':
        applefin.append(row)
for row in englishappsgoogle:
    price = (row[7])
    if price =='0':
        googlefin.append(row)
print(len(applefin))
print(len(googlefin))
```

```python
def freq_table(dataset,index):
    dict={}
    for row in dataset:
        parameter= row[index]
        if parameter in dict:
            dict[parameter]+=1
        else:
            dict[parameter]=0
    return dict

```

```python
def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])
```

```python
prime_genre=[]
for key in freq_table(applefin,11):
    prime_genre.append(key)
```

```python
for genre in prime_genre:
    total =0
    len_genre=0
    for row in applefin:
        rating=float(row[5])
        gn = row[11]
        if gn == genre:
            total+=rating
            len_genre+=1
    print(genre,total / len_genre)

```

```python
categories_android = freq_table(googlefin, 1)
mm={}
for category in categories_android:
    total = 0
    len_category=0
    for row in googlefin:
        catap=row[1]
        if catap==category:
            inst=row[5]
            if '+' in inst:
              inst=(row[5][:-1])
            inst = inst.replace(',','')
            inst = float(inst)
            total+=inst
            len_category+=1
    mm[category]=total/len_category
dict = dict(sorted(mm.items(),key=lambda item:item[1]))
for el in dict:
    print(el,dict[el])


```
