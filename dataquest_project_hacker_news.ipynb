I've made separate lists: 'show' posts,'ask' posts, 'other' posts.I've found average numbers of comments of 'show' and 'ask' posts.  The analysis shows that 'ask' posts receieves more comments on average.The analysis identified the top five hours during which posts receive the highest average number of comments. These peak hours are 15:00, 02:00, 20:00, 16:00, and 21:00

```python
from csv import reader
file = open('hacker_news.csv')
hn=list(reader(file))
file.close()
headers = hn[0]
hn = hn[1:]
```

```python
ask_posts=[]
show_posts=[]
other_posts=[]
for row in hn:
    title = row[1]
    if (title.lower()).startswith('ask hn'):
        ask_posts.append(row)
    elif (title.lower()).startswith('show hn'):
        show_posts.append(row)
    else:
        other_posts.append(row)

```

```python
def avgcomm(sp):
    total=0
    for row in sp:
        total+=int(row[4])
    return total/(len(sp))
avg_ask_comments=avgcomm(ask_posts)
avg_show_comments=avgcomm(show_posts)
print(avg_ask_comments,avg_show_comments)

```

```python
import datetime as dt
result_list=[]
for row in ask_posts:
    created_at=row[6]
    numofcomm=int(row[4])
    result_list.append([created_at,numofcomm])
counts_by_hour={}
comments_by_hour={}
for row in result_list:
    ppp=dt.datetime.strptime(row[0], '%m/%d/%Y %H:%M' )
    hr=(ppp.hour)

    if hr in counts_by_hour:
        counts_by_hour[hr]+=1
        comments_by_hour[hr]+=row[1]
    else:
        counts_by_hour[hr]=1
        comments_by_hour[hr]=row[1]
```

```python
avg_by_hour=[]
for el in (counts_by_hour):
    tot = comments_by_hour[el]/counts_by_hour[el]
    avg_by_hour.append([el,tot])
print((avg_by_hour))
```

```python
swap_avg_by_hour=[]
for row in avg_by_hour:
    swap_avg_by_hour.append([row[1],row[0]])
sorted_swap=sorted(swap_avg_by_hour,reverse = True)
print("Top 5 Hours for Ask Posts Comments")
for row in sorted_swap[:5]:
    print('{}:00: {cr:.2f} average comments per post'.format(row[1],cr=row[0]))
```
